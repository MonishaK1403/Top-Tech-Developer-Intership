```python
# Import NumPy
import numpy as np

# 1. Array Creation
a = np.array([1, 2, 3])           # Create 1D array
b = np.array([[1, 2], [3, 4]])    # Create 2D array
z = np.zeros((2, 3))              # Array of zeros
o = np.ones((3, 3))               # Array of ones
r = np.arange(0, 10, 2)           # Even numbers 0‚Äì8
l = np.linspace(0, 1, 5)          # 5 values between 0‚Äì1
```

```python
# 2. Array Attributes
print(a.shape)     # Shape (rows, columns)
print(a.ndim)      # Number of dimensions
print(a.dtype)     # Data type of elements
```

```python
# 3Ô∏è‚É£ Indexing & Slicing
print(b[0, 1])     # Access element (row 0, col 1)
print(r[1:4])      # Slice from index 1 to 3
```

```python
# 4Ô∏è‚É£ Mathematical Operations
print(a + 5)       # Add scalar
print(a * 2)       # Multiply each element by 2
print(a + b[0])    # Element-wise addition (broadcasting)
print(np.dot(b, b))# Matrix multiplication
```

```python
# 5Ô∏è‚É£ Useful Functions
print(np.sum(a))   # Sum of all elements
print(np.mean(a))  # Average
print(np.max(a))   # Maximum value
print(np.sqrt(a))  # Square root of each element
```

```python
# 6Ô∏è‚É£ Reshaping Arrays
x = np.arange(6).reshape(2, 3)    # Convert 1D to 2x3 array
print(x.T)                        # Transpose rows and columns
```

```python
# 7Ô∏è‚É£ Random Numbers
rand_arr = np.random.rand(2, 3)   # Random floats (0 to 1)
rand_int = np.random.randint(1, 10, 5)  # 5 random ints between 1‚Äì9
```

```python
# 9Ô∏è‚É£ Linear Algebra
m = np.array([[1, 2], [3, 4]])
print(np.linalg.inv(m))           # Inverse
print(np.linalg.det(m))           # Determinant
```

```python
# üîü Logical & Sorting
arr = np.array([1, 3, 2, 3])
print(np.unique(arr))             # Unique elements
print(np.sort(arr))               # Sorted array
print(np.where(arr > 2))          # Indices where condition true
```
