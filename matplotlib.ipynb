```python
#1. Line Graph
#Used to show trends over time
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 8, 12, 20]

plt.plot(x, y)
plt.title("Line Graph Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
#2. Bar Chart
#Used for category comparison
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D"]
values = [25, 40, 30, 20]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
#3. Horizontal Bar Chart
import matplotlib.pyplot as plt

categories = ["Python", "Java", "C", "JS"]
values = [80, 65, 50, 70]

plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.xlabel("Score")
plt.show()
```

```python
#4. Pie Chart
#Used for percentage distribution
import matplotlib.pyplot as plt

labels = ["Food", "Rent", "Travel", "Savings"]
sizes = [30, 40, 15, 15]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart Example")
plt.show()
```

```python
#5. Scatter Plot
#Used to show relationship between two variables
import matplotlib.pyplot as plt

x = [5, 10, 15, 20, 25]
y = [30, 25, 40, 35, 50]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.show()
```

```python
#6. Histogram
#Used to show data distribution
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40, 50]

plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#7. Box Plot
#Used to detect outliers and data spread
import matplotlib.pyplot as plt

data = [10, 12, 14, 15, 18, 20, 25, 100]

plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```

```python
#8. Area Plot
#Used to show cumulative data
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 5, 2, 6, 4]

plt.fill_between(x, y)
plt.title("Area Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
